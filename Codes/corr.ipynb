```python
df = data.dropna()
name = df.columns.values.tolist() 
for i in name:
    df['log'+i] = np.log(df[i])
```

```python
dflog = df
for i in name:
    dflog = dflog.drop(columns=[i], axis=1)
re = dflog.diff()
re.columns = name
```

```python
new=re[['Crude Oil WTI','Natural Gas','Gasoline RBOB','Ethanol','Coal',
        'US Corn','US Wheat', 'US Soybean Meal','US Soybean Oil','US Soybeans', 'Lean Hogs','Live Cattle',
        'Gold','Silver','Platinum','Copper']]
corr = new.corr()
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
plt.subplots(figsize=(12, 12)) # 
sns_plot = sns.heatmap(corr, annot=True, vmin=0, square=True, cmap="Blues")
fig = sns_plot.get_figure()
fig.savefig('corr.jpg') 
```
